BFS(breadth-first search)是图算法的一种，可用与找出两样东西的“最短距离”

**图是什么**
图由节点和边组成，节点可能与众多节点连接，这些节点成为邻居

**寻找芒果经销商**
朋友关系的图

python代码表示

```python
graph = {}
graph["you"] = ["alice", "bob", "claire"]
graph["bob"] = ["anuj", "peggy"]
graph["alice"] = ["peggy"]
graph["claire"] = ["thom", "jonny"]
graph["anuj"] = []
graph["peggy"] = []
graph["thom"] = []
graph["jonny"] = []
```

实现算法

```python
from collections import deque
def breadth_first_search(name):
    search_deque = deque()
    search_deque += graph[name]
    checked = []
    
    while search_deque:
        person = search_deque.pop()
        if person not in checked:
            if person[-1] == 'm':
                return person
        if graph[person]:
                search_deque += graph[person]
breadth_first_search('you')
```

**运行时间**
广度优先搜索的运行时间为 O(人数 + 边数)，这通常写作O(V + E)，其中V为顶点(vertice)数，E为边数。

队列 - 先进先出(FIFO)

栈是 - 后进先出(LIFO)
